I will analyze a report from the World Health Organization on the burden of tuberculosis by countries. I will be converting the untidy report to a tidy dataset to make a clear observation of the variables and the values. After transforming the tidy dataset into a DataFrame in pandas, I will plot the dataset with seaborn for visualization.
 

```python
%matplotlib inline
# import matplotlib for graphics
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages
# import seaborn for graphics
import seaborn as sns
import numpy as np
import pandas as pd
#set some pandas options controling output format
pd.set_option('display.notebook_repr_html',True) # output as flat text and not HTML
pd.set_option('display.max_rows', None) # this is the maximum number of rows we will display
pd.set_option('display.max_columns', None) # this is the maximum number of rows we will display
```

```python
dftb = pd.read_csv(r'c:\data\TB_burden_countries.csv', \
                        header = None, \
                        usecols = [1, 4, 5, 6, 7, 13, 22], \
                        names = ['country', 'region', 'year', 'population', 'incident of TB per 100k population', 'tbhiv_prc test', 'estimated _mort_tbhiv_100k'])
```

```python
dftb.head(5)

```

```python
# Create a scatterplot that shows the relationship between population and comparison of PCR with the routine procedure for diagnosis of tuberculosis in a population with High prevalences of tuberculosis and HIV

# Set the seaborn grid type to whitegrid. Whitegrid displays sightlines, font_scale set the font size.
sns.set(style='whitegrid',font_scale=4)

# Use pastel colors for the plot
sns.set_color_codes('pastel')

# Create a scatterplot that shows the relationship between incidence of tuberculosis per 100k people and people with HIV 
sns.lmplot('population', 'tbhiv_prc test', data=dftb, size=30, aspect=1.294, scatter_kws={"s": 1200}).savefig('C:\Output\population v HIV.pdf')
```

# Conclusion:

This graph present the case of Tuberculosis in detection rate per 100,000. The cases associated with HIV patients have contibuted to a high risk of those patients also developing Tuberculosis. Developing countries have faced an epedic and an increase of "78 per 100,000 in 1991 to 581 per 100,000 in 2001, with a total number of 12,963 cases".

```python
# Create a scatterplot that shows the relationship between years and population infected with TB
sns.lmplot('year', 'population', data=dftb, size=30, aspect=1.294, scatter_kws={"s": 1200}).savefig('C:\Output\year v populaion.pdf')
```

```python
# Set the seaborn grid type to whitegrid. Whitegrid displays sightlines, font_scale set the font size.
sns.set(style='whitegrid',font_scale=4)

# Use pleasent pastel colors in our plot
sns.set_color_codes('pastel')

# change the plotsize to 11" x 3, by 8.5" * 3
fig, ax = plt.subplots(figsize=(33, 25.5))
# display a barchart that shows the estimated mortality of TB cases (all forms, excluding HIV) per 100 000 population..
sns.barplot(x='estimated _mort_tbhiv_100k', y='tbhiv_prc test', data=dftb, label='Estimated mortality rate by TB_HIV per 100k', color='b')
fig.savefig('C:\Output\Mortality Rate and TB_HIV.pdf')
```
