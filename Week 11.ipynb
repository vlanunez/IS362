```python
# this is a simple funtion that takes some basic arguments for the New York Times API and returns the JSON respnse
def NYTSearch(strQuery, strStartDate, strEndDate):
    # Set the URL for the API
    url = 'https://api.nytimes.com/svc/search/v2/articlesearch.json'
    dictParams = {'api-key': "76GgTJdfNpk359GHXWYWqCF2F2O0AAkq",
                 'q': strQuery,
                 'begin_date': strStartDate,
                 'end_date': strEndDate,
                 'page': 1}
    

    return(requests.get(url, params = dictParams))
```

```python
# send the request to the New York Times API
response = NYTSearch("cats and dogs", 20160101, 20170101) 
```
